```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
df4=df1[['Country Code']] = df1[['Country Code']].apply(lambda col:pd.Categorical(col).codes)
```

```python
df1
```

```python
scaler = MinMaxScaler()
norm_data = df1.copy()
def minmaxscaler(x):
    for IncomeGroup, SpecialNotes in x.iteritems():
        x[IncomeGroup] = scaler.fit_transform(np.array(SpecialNotes).reshape(-1, 1))
    
minmaxscaler(norm_data)
norm_data.head()
```

```python
kmc = list(range(3,5))
sum_of_squared_distances = []
for i in kmc:
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(norm_data)
    sum_of_squared_distances.append(kmeans.inertia_)

plt.figure(figsize=(9, 10))
plt.plot(kmc, sum_of_squared_distances, 'go--')
plt.xlabel('Number of Clusters')
plt.ylabel('Within Cluster Sum of squares')
plt.title('Elbow Curve to find optimum K')
```

```python
# Instantiating
kmeans4 = KMeans(n_clusters = 5)

# Training the model
kmeans4.fit(norm_data)

# predicting
y_pred = kmeans4.fit_predict(norm_data)
print(y_pred)

df1
```

```python
df1['IncomeGroup'] = y_pred+1
```

```python
df1
```

```python
centroids = kmeans4.cluster_centers_
centroids = pd.DataFrame(centroids, columns=['Country Code','Region','IncomeGroup','SpecialNotes','TableName'])
centroids.index = np.arange(1, len(centroids)+1)
centroids
```
